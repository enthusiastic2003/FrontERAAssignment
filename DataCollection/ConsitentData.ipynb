```python
# Consitent data path
import os
consistent_data_path = "./dataClassified/cry_data/"

if not os.path.exists(consistent_data_path):
    os.makedirs(consistent_data_path)

```

```python
## DonateACRY Datasets

infant_cry = "./tempData/1/donateacry_corpus/"

# Find folders inside the infant_cry folder
import os
folders = os.listdir(infant_cry)
print(folders)

# Create a list of file paths to all the .wav files inside the folders

file_paths = []
save_file_paths = []
counter = 0
for folder in folders:
    files = os.listdir(infant_cry + folder)
    for file in files:
        if file.endswith(".wav"):
            file_paths.append(infant_cry + folder + "/" + file)
            save_file_paths.append( consistent_data_path+ f"donateacry_{counter}.wav")
            counter += 1
```

```python
import soundfile as sf
import librosa
import numpy as np
import os

# Desired settings
target_sr = 16000  # 16kHz

for i,file_path in enumerate(file_paths):
    # Load audio with librosa (handles resampling and mono conversion)
    data, _ = librosa.load(file_path, sr=target_sr, mono=True)
    
    # Ensure float32 type
    data = data.astype(np.float32)

    # Define output path
    output_path = os.path.join(consistent_data_path, f"donateacry_{i}.wav")
    
    # Write as 16-bit PCM WAV
    sf.write(output_path, data, target_sr, subtype='PCM_16')

    print(f"Processed and saved: {output_path}")
```

```python
# Load Mendelyan Cry Datasets

mendelyan_cry = "./tempData/MendelyCryData/Infants Cry Sound/Dataset"

# Find folders inside the mendelyan_cry folder
folders = os.listdir(mendelyan_cry)
print(folders)
```

```python
# Create a list of file paths to all the .wav files inside the folders

file_paths = []
save_file_paths = []
counter = 0
for folder in folders:
    files = os.listdir(mendelyan_cry + "/" + folder)
    for file in files:
        if file.endswith(".wav") and counter<440:
            file_paths.append(mendelyan_cry + "/" + folder + "/" + file)
            save_file_paths.append(consistent_data_path + "mendely_cry_" + str(counter) + ".wav")
            counter += 1

# Consitent data path

consistent_data_path = "./dataClassified/cry_data/"

if not os.path.exists(consistent_data_path):
    os.makedirs(consistent_data_path)

```

```python
import librosa
import soundfile as sf
import numpy as np
```

```python

# Desired settings
target_sr = 16000  # 16kHz

for i,file_path in enumerate(file_paths):
    # Load audio with librosa (handles resampling and mono conversion)
    data, _ = librosa.load(file_path, sr=target_sr, mono=True)
    
    # Ensure float32 type
    data = data / np.max(np.abs(data))
    data = data.astype(np.float32)

    # Define output path
    output_path = os.path.join(consistent_data_path, os.path.basename(save_file_paths[i]))
    
    # Write as 16-bit PCM WAV
    sf.write(output_path, data, target_sr, subtype='PCM_16')

    print(f"Processed and saved: {output_path}")
```
